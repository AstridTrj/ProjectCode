```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
ladder = pd.read_csv('E:\\what-study\\project\\ladder\\ladder.csv')
```

```python
f, ax = plt.subplots(figsize=(8, 27))
sns.barplot(x="score", y="name", data=ladder, color="#555555", ci=0)
plt.title('CPU')
plt.show()
```

```python
main_type = {'OPPO', 'vivo', '', '', '', '', 
            '', '', '', '', '', '', ''}
for i in main_type:
    f = open('E:\\what-study\\project\\rank\\' + i + '.csv', encoding='utf-8')
    hot_phone = pd.read_csv(f)
    if '' in hot_phone.columns:
        hot_phone[''] = hot_phone[''].map(lambda x: x[:x.find('%')])
        hot_phone[''] = hot_phone[''].map(lambda x: x[:x.find('')])
        hot_phone[''] = hot_phone[''].astype(float)
        f, ax = plt.subplots(figsize=(8, 15))
        sns.barplot(x=hot_phone[''], y=hot_phone[''], data=hot_phone, color="#612563", ci=0)
        plt.title(i + '')
        plt.xlabel('(%)')
        plt.show()
```
